Download all of the data for D. rerio from PsicQuic using the following command:

```python
import requests

base_url = "http://www.ebi.ac.uk/Tools/webservices/psicquic/intact/webservices/current/search/query/Danio%20Rerio"
output_filename = "zfish_psicquic_results.xml"  # Output filename for all results

# Define the number of results per request and the total number of results you want
results_per_request = 100
total_results = 7500  # Change this to your desired total number of results

with open(output_filename, "wb") as combined_xml_file:
    for start_result in range(0, total_results, results_per_request):
        url = f"{base_url}?format=xml25&firstResult={start_result}&maxResults={results_per_request}"

        try:
            # Send an HTTP GET request to the URL
            response = requests.get(url)

            if response.status_code == 200:
                # Check if the response content is in XML format (optional)
                if "xml" in response.headers.get("content-type", "").lower():
                    # Append the content to the combined XML file
                    combined_xml_file.write(response.content)
                    print(f"XML content for {start_result}-{start_result + results_per_request} appended to {output_filename}")
                else:
                    print("The content is not in XML format.")
            else:
                print(f"Failed to retrieve the content for {start_result}-{start_result + results_per_request}. Status code: {response.status_code}")
        except requests.exceptions.RequestException as e:
            print(f"An error occurred for {start_result}-{start_result + results_per_request}: {e}")
```
